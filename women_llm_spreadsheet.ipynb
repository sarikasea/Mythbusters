```python
import pandas as pd
import google.generativeai as genai
import json
import time

# Authenticate Gemini
genai.configure(api_key=userdata.get('GEMINI_API_KEY'))

# Load your table from Google Sheets as DataFrame (as done in your notebook)
# dftrans = get_as_dataframe(sheet)  # Already loaded

# Specify columns to fill (all those you confirmed earlier)
columns_to_fill = [
    'Date of Birth', 'Date of Death', 'Ethnicity', 'Religion', 'Languages',
    'Education', 'Primary Occupation', 'Cause of Death', 'Primary Challenge',
    'Affiliations', 'Awards', 'Key Works', 'Contemporaries', 'Legacy Today',
    'Film/TV References', 'Wikipedia URL', 'Imagery', 'Social Class',
    'Sexuality', 'Barrier Broken', 'Migration', 'Mentorship'
]

# Create a Gemini model instance only once
model = genai.GenerativeModel('gemini-1.5-flash')  # Or 'gemini-1.5-pro'

def get_llm_response(row, columns):
    prompt = f"""
    Given the following information about the historical figure "{row['Name']}", fill in as many details as possible for these columns:
    Existing: {row.to_dict()}
    Needed: {columns}
    Respond as JSON with only the new fields and values.
    """
    response = model.generate_content(prompt)
    try:
        # Correctly extract the JSON part from the response
        text_response = response.text.strip()
        # Find the start and end of the JSON block
        json_start = text_response.find('{')
        json_end = text_response.rfind('}') + 1
        if json_start != -1 and json_end != -1:
            json_str = text_response[json_start:json_end]
            data = json.loads(json_str)
        else:
            print(f"No JSON object found in the response for {row['Name']}")
            data = {}
    except Exception as e:
        print(f"Error parsing Gemini response for {row['Name']}: {e}")
        data = {}
    return data

# Process and fill the missing fields in the first row of your DataFrame
if not dftrans.empty:
    idx = 0
    row = dftrans.iloc[idx]
    missing = [col for col in columns_to_fill if pd.isnull(row.get(col, '')) or str(row.get(col, '')).strip() == '']
    if missing:  # Only if there is anything missing
        vals = get_llm_response(row, missing)
        for col, val in vals.items():
            if col in dftrans.columns:
                dftrans.at[idx, col] = val
        time.sleep(1)  # Respect API rate limits; adjust as needed

# Display the updated first row
display(dftrans.head(1))

# Export back to CSV and/or Excel
dftrans.to_csv('Historical-Women-Completed.csv', index=False)
```
