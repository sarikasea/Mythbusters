# Column Clean Up

```python
# Example: Define a function to categorize the 'Love Life' column
def categorize_love_life(value):
    lower = str(value).lower()
    if "married" in lower:
        return "Married"
    elif "widow" in lower:
        return "Widowed"
    elif "affair" in lower or "lover" in lower or "romance" in lower:
        return "Affairs/Romance"
    elif "single" in lower or "never married" in lower or "celibate" in lower:
        return "Single/Celibate"
    elif "arranged" in lower:
        return "Arranged Marriage"
    elif "unknown" in lower or pd.isna(value):
        return "Unknown/Private"
    else:
        return "Other"

# Create new column with categorizations
new_unique['Love Life Category'] = new_unique['Love Life'].apply(categorize_love_life)
```

```python
from google.colab import userdata
import google.generativeai as genai
import json
import time

genai.configure(api_key=userdata.get("GEMINI_API_KEY"))

def get_llm_response(row, columns_to_fill, max_retries=3, delay_seconds=2):
    prompt = f"""
You are an AI assistant helping populate missing data fields.

Given this information about "{row['Name']}", fill in the following columns: {columns_to_fill}

Existing data:
{row.to_dict()}

Respond ONLY in valid JSON format like:
{{
  "Field1": "Value1",
  "Field2": "Value2"
}}

Do not include explanations or formatting — only output valid JSON.
    """

    for attempt in range(1, max_retries + 1):
        try:
            model = genai.GenerativeModel('gemini-1.5-flash')
            response = model.generate_content(prompt)

            print(f"📤 Attempt {attempt} - Prompt sent to Gemini:\n", prompt)
            print("\n📥 Gemini raw response:\n", repr(response.text))

            data = json.loads(response.text)
            return data  # ✅ Success
        except Exception as e:
            print(f"\n⚠️ Attempt {attempt} failed - Error during LLM call or JSON parsing:", e)
            print("❌ Raw response text:\n", repr(response.text if 'response' in locals() else 'No response'))

            if attempt < max_retries:
                print(f"🔁 Retrying in {delay_seconds} seconds...\n")
                time.sleep(delay_seconds)
            else:
                print("❌ Max retries reached. Skipping this row.\n")
                return {}  # Final fallback

    return {}
```
